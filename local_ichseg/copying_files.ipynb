```python
import shutil
import os
def copy_files(file, source_dir, dest_dir):
    org_annotations = pd.read_csv(file)
    df = pd.DataFrame(org_annotations)
    cols_to_check = [col for col in df.columns if col != 'Filename']
    df_subset = df[df[cols_to_check].eq('yes').any(axis=1)]
    if not os.path.exists(dest_dir):
        os.makedirs(dest_dir)
    for index, row in df_subset.iterrows():
        filename = row['Filename']
        source_path = f"{source_dir}/{filename}"
        dest_path = f"{dest_dir}/{filename}"
        try:
            shutil.copy(source_path, dest_path)
        except Exception as e:
            print(f"Error copying {filename}: {e}")
```

```python
copy_files('/Users/rushil/ichseg/HD_CTBET/annotations_hdctbet_combined.csv',
           '/Users/rushil/ichseg/HD_CTBET/image_ss_hdctbet',
           '/Users/rushil/ichseg/HD_CTBET/image_ss_hdctbet_failures')
```
